# Converting Numpy Arrays to Pandas DataFrames for Sensor and Position Data

```python
X = sensors_data
y = position_data
```

```python
# convert numpy array into pandas dataframe
X = pd.DataFrame(X)
y = pd.DataFrame(y)
```

# 1. Importing Deep Learning Libraries

```python
import time
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error
from keras.layers import LSTM, BatchNormalization, Activation, Dense, Dropout
```

# 2. Define Network with KFold Cross Validation

```python
# Perform 5-fold cross-validation
kfold = KFold(n_splits=5, shuffle=True)
mse_scores = []
mae_scores = []
rmse_scores = []
time_taken = []
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True)

for fold, (train_index, test_index) in enumerate(kfold.split(X), 1):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model = Sequential()
    
    model.add(LSTM(512, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(BatchNormalization())
    model.add(Activation('relu'))

    model.add(LSTM(256, return_sequences=True))
    model.add(LSTM(128))
    
    model.add(Dense(64))
    model.add(Dense(3))
    
    adam = Adam(lr=0.0001)
    model.compile(loss='mean_squared_error', optimizer=adam)

    start_time = time.time()
    history = model.fit(
        X_train, y_train,
        epochs=200, batch_size=6,
        validation_data=(X_test, y_test),
        callbacks=[early_stopping]
    )
    end_time = time.time()

    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    mse_scores.append(mse)
    mae_scores.append(mae)
    rmse_scores.append(rmse)
    time_taken.append(end_time - start_time)

    print("Evaluation Results for Fold", fold)
    print("Mean Squared Error (MSE):", mse)
    print("Mean Absolute Error (MAE):", mae)
    print("Root Mean Squared Error (RMSE):", rmse)
    print("Time taken:", end_time - start_time)
    print()   
```

```python
model.summary()
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model,'LSTM.png', show_shapes = True)
```

# 3. Evaluate Network: Calculate average results across all folds

```python
avg_mse = np.mean(mse_scores)
avg_mae = np.mean(mae_scores)
avg_rmse = np.mean(rmse_scores)
avg_time_taken = np.mean(time_taken)
```

# 4. Create a DataFrame for all fold results and average results

```python
results_df = pd.DataFrame({
    'Fold': list(range(1, kfold.n_splits + 1)),
    'MSE': mse_scores,
    'MAE': mae_scores,
    'RMSE': rmse_scores,
    'Time taken': time_taken
})

# Append average results to the DataFrame
avg_results = {'Fold': 'Average', 'MSE': avg_mse, 'MAE': avg_mae, 'RMSE': avg_rmse, 'Time taken': avg_time_taken}
results_df = results_df.append(avg_results, ignore_index=True)
```

# 5. Save the results to an Excel file

```python
# Save the results to an Excel file
results_df.to_excel('Sensors 16_PL_model_2_Scattered_iReg_f.xlsx', index=False)

# Display the results table
print("Results for all Folds and Average Results:")
print(results_df)


print("Results saved to 'Sensors 16_PL_model_2_Scattered_iReg_f.xlsx'")
```
